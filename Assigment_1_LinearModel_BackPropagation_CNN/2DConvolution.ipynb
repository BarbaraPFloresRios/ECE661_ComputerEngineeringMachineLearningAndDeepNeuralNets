```python
import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
from io import BytesIO
import requests
from PIL import Image
```

```python

url = "https://raw.githubusercontent.com/BarbaraPFloresRios/ECE661_ComputerEngineeringMachineLearningAndDeepNeuralNets/main/Assigment_1_LinearModel_BackPropagation_CNN/example_image.JPG"
response = requests.get(url)
img = Image.open(BytesIO(response.content)).convert("L")  

img_array = np.array(img)

kernel = np.array([
    [0, -0.5, 0],
    [-0.5, 1, -0.5],
    [0, -0.5, 0]
])

output_image = convolve2d(img_array, kernel, mode='same', boundary='symm')

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Filtered Image with Kernel")
plt.imshow(output_image, cmap="gray")
plt.axis("off")

plt.show()
```
